## 1. What is a Variable?

A variable in Python is a name used to store a value.

Variables are useful in healthcare analytics because they make it easier to store, reference, reuse, and analyze values such as patient counts, fall incidents, compliance rates, readmission counts, and quality indicators.

```python
patient_group = "Older Adults"
total_patients = 150
fall_incidents = 12

print(patient_group)
print(total_patients)
print(fall_incidents)
```

## 2. Common Python Data Types

| Data Type | Meaning | Example |
|---|---|---|
| String | Text | `"Fall Risk Dashboard"` |
| Integer | Whole number | `150` |
| Float | Decimal number | `94.5` |
| Boolean | True or False | `True` |
| List | Collection of values | `["Falls", "Readmissions", "Infections"]` |

## 3. String

A string is text. Strings are written inside quotation marks.

```python
dashboard_name = "Fall Risk Dashboard"
care_area = "Long-Term Care"

print("Dashboard Name:", dashboard_name)
print("Care Area:", care_area)
```

## 4. Integer

An integer is a whole number.

```python
total_patients = 150
fall_incidents = 12

print("Total Patients:", total_patients)
print("Fall Incidents:", fall_incidents)
```

## 5. Float

A float is a number with decimals.

```python
compliance_rate = 94.5
average_age = 78.6

print("Compliance Rate:", compliance_rate)
print("Average Age:", average_age)
```

## 6. Boolean

A Boolean value is either `True` or `False`.

This is useful for yes/no or true/false healthcare data.

```python
training_completed = True
high_risk_patient = False

print("Training Completed:", training_completed)
print("High Risk Patient:", high_risk_patient)
```

## 7. List

A list stores multiple values in one variable.

```python
quality_indicators = ["Falls", "Readmissions", "Infections", "Medication Errors"]

print("Quality Indicators:", quality_indicators)
```

## 8. Comments

Comments are notes in Python code. They start with `#`.

Python ignores comments when the code runs.

```python
# This stores the name of the dashboard
dashboard_name = "Fall Risk Dashboard"

# This stores the number of fall incidents
fall_incidents = 12

print(dashboard_name)
print(fall_incidents)
```

## 9. Healthcare Analytics Example

Python variables can be used to store healthcare analytics values such as dashboard name, care area, reporting period, patient count, fall incidents, readmission count, compliance rate, and quality indicators.

```python
dashboard_name = "Fall Risk Dashboard"
care_area = "Long-Term Care"
reporting_period = "2025/2026"
total_patients = 150
fall_incidents = 12
readmission_count = 8
compliance_rate = 94.5
average_age = 78.6
bed_occupancy_rate = 91.3
patient_satisfaction_score = 87.2
training_completed = True
high_risk_patient = False
monthly_incident_reviews_completed = True
quality_indicators = ["Falls", "Readmissions", "Infections", "Medication Errors"]

print("Dashboard Name:", dashboard_name)
print("Care Area:", care_area)
print("Reporting Period:", reporting_period)
print("Total Patients:", total_patients)
print("Fall Incidents:", fall_incidents)
print("Readmission Count:", readmission_count)
print("Compliance Rate:", compliance_rate)
print("Average Age:", average_age)
print("Bed Occupancy Rate:", bed_occupancy_rate)
print("Patient Satisfaction Score:", patient_satisfaction_score)
print("Training Completed:", training_completed)
print("High Risk Patient:", high_risk_patient)
print("Monthly Incident Reviews Completed:", monthly_incident_reviews_completed)
print("Quality Indicators:", quality_indicators)
```
