# Lesson 04: Lists and Dictionaries in Python

In this lesson, we will learn how to store multiple related values using **lists** and **dictionaries**.

These concepts are important in healthcare analytics because real-world data often includes multiple patients, units, quality indicators, metrics, and reporting periods.

## 1. What is a List?

A **list** stores multiple values in one variable.

For example, a list can store multiple healthcare quality indicators.

````python
# Create a list of healthcare quality indicators
quality_indicators = ["Falls", "Readmissions", "Infections", "Medication Errors"]

print("Quality Indicators:", quality_indicators)
````

## 2. Accessing Items in a List

Python starts counting from **0**, not 1.

That means the first item in a list is at position `0`.

````python
# Access items in the list
print("First Quality Indicator:", quality_indicators[0])
print("Second Quality Indicator:", quality_indicators[1])
````

## 3. Adding Items to a List

Use `.append()` to add a new item to the end of a list.

````python
# Add a new quality indicator
quality_indicators.append("Pressure Injuries")

print("Updated Quality Indicators:", quality_indicators)
````

## 4. Counting Items in a List

Use `len()` to count how many items are in a list.

````python
# Count the number of quality indicators
print("Number of Quality Indicators:", len(quality_indicators))
````

## 5. What is a Dictionary?

A **dictionary** stores data as **key-value pairs**.

A dictionary is useful when you want to store related information about one thing, such as a patient group, unit, dashboard, or metric summary.

````python
# Create a dictionary with healthcare summary values
patient_summary = {
    "patient_group": "Older Adults",
    "total_patients": 150,
    "fall_incidents": 12,
    "readmission_count": 8,
    "compliance_rate": 94.5
}

print("Patient Summary:", patient_summary)
````

## 6. Accessing Dictionary Values

You can access a dictionary value by using its key.

````python
# Access values from the dictionary
print("Patient Group:", patient_summary["patient_group"])
print("Total Patients:", patient_summary["total_patients"])
print("Fall Incidents:", patient_summary["fall_incidents"])
print("Compliance Rate:", patient_summary["compliance_rate"])
````

## 7. Updating Dictionary Values

You can update an existing value in a dictionary.

````python
# Update the compliance rate
patient_summary["compliance_rate"] = 96.2

print("Updated Compliance Rate:", patient_summary["compliance_rate"])
````

## 8. Adding New Key-Value Pairs

You can also add new information to a dictionary.

````python
# Add a new key-value pair
patient_summary["bed_occupancy_rate"] = 91.3

print("Bed Occupancy Rate:", patient_summary["bed_occupancy_rate"])
print("Updated Patient Summary:", patient_summary)
````

## 9. List of Dictionaries

A list can contain multiple dictionaries.

This is very useful because many datasets are like rows in a table.

| Unit | Fall Incidents | Compliance Rate |
|---|---:|---:|
| Unit A | 5 | 91.2 |
| Unit B | 3 | 95.4 |
| Unit C | 7 | 88.9 |

````python
# Create a list of dictionaries
unit_metrics = [
    {"unit": "Unit A", "fall_incidents": 5, "compliance_rate": 91.2},
    {"unit": "Unit B", "fall_incidents": 3, "compliance_rate": 95.4},
    {"unit": "Unit C", "fall_incidents": 7, "compliance_rate": 88.9}
]

print("Unit Metrics:", unit_metrics)
````

## 10. Accessing Values from a List of Dictionaries

To access information from a list of dictionaries, first choose the list position, then choose the dictionary key.

````python
# Access values from a list of dictionaries
print("First Unit:", unit_metrics[0]["unit"])
print("First Unit Fall Incidents:", unit_metrics[0]["fall_incidents"])
print("Second Unit Compliance Rate:", unit_metrics[1]["compliance_rate"])
````
